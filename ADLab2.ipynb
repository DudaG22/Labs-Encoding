```python
import pandas as pd
import datetime
```

```python
data_atual = datetime.datetime.now()
```

```python
data_atual
```

```python
# Range de datas dos 20 dias anteriores Ã  data atual

date_list = [data_atual - datetime.timedelta(days = x) for x in range(0,20) ]
```

```python
date_list
```

```python
df = pd.DataFrame(date_list)
```

```python
df.head()
```

```python
df.columns = ['data']
```

```python
df['dia_da_semana'] = df.data.dt.day_name()
```

```python
df
```

```python
dict_dia_semana = {'Monday':1, "Tuesday":2, "Wednesday":3, "Thursday":4, "Friday":5, "Saturday":6, "Sunday":7}
```

```python
df['cod_dds'] = df.dia_da_semana.map(dict_dia_semana)
```
